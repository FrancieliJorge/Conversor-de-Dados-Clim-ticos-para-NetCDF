```python
import os
import shutil
```

```python
# Defina os diretÃ³rios principais
source_dir = '/CAMINHO/'  # Substitua pelo caminho da pasta principal
destination_dir = '/CAMINHO/'  # Substitua pelo caminho da pasta de destino
```

```python
# Defina os meses e dias que queremos processar
months = ['01', '02', '03', '11', '12']
days = {
    11: [3, 14],
    12: [1, 15],
    1: [4, 14],
    2: [1, 15],
    3: [3, 14]
}
```

```python
# FunÃ§Ã£o para verificar se uma pasta representa uma data vÃ¡lida
def is_valid_date_folder(folder_name):
    if len(folder_name) != 10:
        return False
    year = folder_name[:4]
    month = folder_name[4:6]
    day = folder_name[6:8]
    hour = folder_name[8:10]
    return (year.isdigit() and month.isdigit() and day.isdigit() and hour.isdigit() and
            int(year) >= 1999 and int(year) <= 2011 and
            int(month) in days and int(day) in days[int(month)] and hour == '00')
```

```python
# Percorrer todas as subpastas na pasta principal
for subdir in os.listdir(source_dir):
    subdir_path = os.path.join(source_dir, subdir)
    if os.path.isdir(subdir_path) and is_valid_date_folder(subdir):
        # Criar a nova pasta de destino
        new_dest_dir = os.path.join(destination_dir, subdir)
        if not os.path.exists(new_dest_dir):
            os.makedirs(new_dest_dir)
        
        # Copiar os arquivos desejados
        file_name = 'ECMWF_PREC_GPCP_{}.nc'.format(subdir)
        src_file_path = os.path.join(subdir_path, file_name)
        dest_file_path = os.path.join(new_dest_dir, file_name)
        if os.path.isfile(src_file_path):
            shutil.copy2(src_file_path, dest_file_path)
            print('Copiado: {}'.format(dest_file_path))
        else:
            print('Arquivo nÃ£o encontrado: {}'.format(src_file_path))
```
