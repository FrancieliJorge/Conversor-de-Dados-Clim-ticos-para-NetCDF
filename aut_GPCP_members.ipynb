```python
from glob import glob
import subprocess
import pandas as pd
import os
from datetime import datetime

```

```python
# Função para gerar as datas específicas
def generate_specific_dates(start_year, end_year):
    valid_days = {
        11: [3, 14],
        12: [1, 15],
        1: [4, 14],
        2: [1, 15],
        3: [3, 14]
    }
    specific_dates = []
    for year in range(start_year, end_year + 1):
        for month, days in valid_days.items():
            for day in days:
                specific_dates.append(datetime(year, month, day, 0))
    return specific_dates

# Função para remover ou comentar a linha OPTIONS _cal
def clean_ctl_file(path):
    with open(path, 'r') as file:
        lines = file.readlines()

    with open(path, 'w') as file:
        for line in lines:
            if "OPTIONS _cal" in line:
                file.write("#" + line)  # Comentar a linha
            else:
                file.write(line)

# Gerar datas de 1999 até 2011
rodadas = generate_specific_dates(1999, 2011)
```

```python
for rod in rodadas:
    paths = sorted(glob(rod.strftime("%Y%m%d%H/*.ctl")))
    for path in paths:
        clean_ctl_file(path)  # Limpar o arquivo .ctl
        
        path_in = path
        path_out = path.replace('.ctl', '.nc')
        
        # Verificar se o arquivo .ctl existe
        if not os.path.isfile(path_in):
            print("Arquivo .ctl não encontrado: {}".format(path_in))
            continue
        
        # Verificar se o arquivo de saída já existe
        if os.path.isfile(path_out):
            print("Arquivo .nc já existe: {}".format(path_out))
            continue
        
        cmd = "cdo -f nc import_binary {path_in} {path_out}".format(path_in=path_in, path_out=path_out)
        print(cmd)
        try:
            subprocess.check_call(cmd, shell=True)
        except subprocess.CalledProcessError as e:
            print("Erro ao executar o comando: {}".format(e))
            continue

```
